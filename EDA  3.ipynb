# Exploring the data

```python
numeric_features=[features for features in df.columns if df[features].dtypes!='O']
categorical_features=[features for features in df.columns if df[features].dtypes=='O']
```

```python
numeric_features
```

```python
categorical_features
```

```python
df['App'].value_counts(normalize=True)
```

```python
len(df['App'].value_counts())
```

```python
num_df=df[numeric_features]
```

```python
cat_df=df[categorical_features]
```

```python
sns.kdeplot(num_df['Rating'])
```

```python
for feature in num_df :
    sns.kdeplot(df[feature])
    plt.xlabel(feature)
    plt.ylabel('count')
    plt.title('numerical features')
    plt.show()
```

```python
plt.figure(figsize=(15 , 15))
plt.suptitle('Univariate analysis of numerical features',fontsize=20,fontweight='bold',alpha=0.8,y=1)

for i in range(0 , len(numeric_features)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=df[numeric_features[i]],shade=True,color='r')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```
